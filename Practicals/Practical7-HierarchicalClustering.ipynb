```python
#Hierarchical clustering using Agglomerative approach
```

```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('diabetes2.csv')
```

```python
data.head(5)
```

```python
x = data.iloc[:,[7,2]].values
```

```python
x
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters = 3,affinity ='euclidean',linkage ='ward')
```

```python
model.fit(x)
```

```python
labels = model.labels_
```

```python
labels
```

```python
plt.scatter(x[labels==0, 0], x[labels==0, 1], s=50, marker='o', color='violet')
plt.scatter(x[labels==1, 0], x[labels==1, 1], s=50, marker='o', color='lightblue')
plt.scatter(x[labels==2, 0], x[labels==2, 1], s=50, marker='o', color='lightgreen')

plt.show()
```
