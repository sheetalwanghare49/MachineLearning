```python
sns.pairplot(data,x_vars = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'],y_vars = 'Outcome',kind = 'scatter', diag_kind = None)
```

```python
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
plt.scatter(data.Age,data['BloodPressure'])
plt.xlabel('Age')
plt.ylabel('BloodPressure')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data[['Age','BloodPressure']])
y_predicted
```

```python
data['cluster']=y_predicted
data.head()
```

```python
km.cluster_centers_

data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
plt.scatter(data1.Age,data1['BloodPressure'],color='lightgreen')
plt.scatter(data2.Age,data2['BloodPressure'],color='pink')
plt.scatter(data3.Age,data3['BloodPressure'],color='lightblue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.legend()
```

```python
#Elbow method
```

```python
x1 = data.iloc[:,[8,3]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss_list=[]
```

```python
for i in range(1,16):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit_predict(x1)
    wcss_list.append(kmeans.inertia_)
```

```python
wcss_list
```

```python
plt.plot(range(1,16),wcss_list)
```
